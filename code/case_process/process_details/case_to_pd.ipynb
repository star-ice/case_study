```python
import pandas as pd
import json
from IPython.display import display, HTML
import os
from io import StringIO
from urllib.request import urlopen
import json
import requests
pd.reset_option('display.max_rows', 500)
```

```python
def row_join(df):
    return'\ '.join(df['text'].values)
```

```python
def process_one_json(data):
    df_json = pd.read_json(StringIO(data))
#     display(df_json)
    df_seg = pd.DataFrame(df_json.returnData.segments)
    df_temp = pd.DataFrame(df_json.returnData[['id', 'caseTitle', 'court', 'judgementDate', 'procedure', 'reason']])
    new_df_temp = df_temp.unstack()
    new_df_temp = new_df_temp.unstack().reset_index()
    df_seg_temp = df_seg.groupby('type').apply(row_join).reset_index()
    df_seg_temp = df_seg_temp.rename(columns={0:'text'})
    new_df_seg_temp = pd.DataFrame([list(df_seg_temp.text)], columns=list(df_seg_temp.type))
    new_df_json = pd.concat([new_df_temp, new_df_seg_temp], axis=1)
    new_df_json = new_df_json.drop(columns=['index'])
#     display(new_df_json)
    return new_df_json
    
```

```python
def batch_process_jsons(url):
    if not os.path.exists(url):
        print("路径不存在！", url)
        return
    batch_df_json = pd.DataFrame()
#     display(batch_df_json)
    files = os.listdir(url)
    for file in files:
        if os.path.splitext(file)[-1][1:] != "json": continue
        if not os.path.isdir(file): #判断是否是文件夹，不是文件夹才打开
            with open(url + "/" + file, 'r', encoding="UTF-8") as f: 
                data = json.load(f)
            batch_df_json = batch_df_json.append(process_one_json(data))
            
    return batch_df_json[['id', 'caseTitle', 'court', 'judgementDate', 'procedure', 'reason', 'CASE_NO', \
                          'COURT_HELD', 'JUDGE', 'JUDGEMENT_RESULT', 'LITIGANT', \
                          'TRIAL_PROCESS']]
```

<h3>Process all case files into csv</h3>

```python
def merge_dfs(df, original_df):
    original_df.reset_index(drop=True, inplace=True)
    original_df = original_df.add_prefix('ORIGINAL_')
    new_df = pd.concat([df, original_df], axis=1)
    final_df = new_df[['case_id', 'party_info', 'judgement_date', 'procedure', 'reason',
       'court_name', 'judge', 'judgere_keyword', 
       'judgement_result']]
    final_df['is_party_info_incomplete'] = final_df['party_info'].apply(lambda x: True if "{}" in str(x) else False)
    final_df['is_judgere_incomplete'] = final_df['judgement_result'].apply(lambda x: True if "[]" in str(x) else False)
#     display(final_df[['judgement_result', 'is_judgere_incomplete', 'ORIGINAL_JUDGEMENT_RESULT']])
    return final_df
```

```python
# Add success rate
cost_type1 = ['受理费', '诉讼费', '上诉费', '减半', '收取'] #用作判断一审赢率

cost_type2 = ['货款', '价款', '购物款', '购酒款', '购药款', '购货款', '餐费', '餐饮费'] #不同商品的价格
judge_verb2 = ["退还", "返还", "退款"]

cost_type3 = ['赔偿金', '赔偿款', '统一赔偿费用', '十倍价款', '十倍货款', '价款的十倍', '货款的十倍'] #通常是商品价格的十倍左右
judge_verb3 = ["支付", "赔偿", "赔付", "迳付", "给付"]

cost_type4 = ['公证费', '公告费', '运费', '交通费', '误工费', '打印费', '鉴定费', '邮寄费', '医疗费', \
             '住宿费', '产品质量监督检验费', '其他费用', '管辖权异议费', '检测费'] #案件附加费用

# cost_type5 = ['共计', '合计']#可能是多个类型费用的总和，比较混淆

cost_types = ['cost_type1', 'cost_type2', 'cost_type3', 'cost_type4', 'unknown_cost_type']
```

```python
#因为ipynb notebook的导入有问题，所以把success_rate notebook中的内容搬到了这里
def merge_dict(x,y):
    for k,v in x.items():
                if k in y.keys():
                    y[k] += v
                else:
                    y[k] = v
    return y

def read_judgement_result(judgement_result):
    '''
    judgement_result组合：
    1. ['cost_type', '', '50元', ['吴明德']]
    2. ['judge_verb', '', '50元', ['吴明德']]
    3. ['judge_verb', 'cost_type', '50元', ['吴明德']]
    4. ['cost_type', 'cost_type', '50元', ['吴明德']]
    5. ['judge_verb', '', '50元', ['吴明德']]
    6. ['cost_type', 'unknown', '50元', ['吴明德']]
    7. ['unknown', '', '50元', ['吴明德']]
    先判断list中第二个元素是否在cost_type中，如果在，则将该整体list内容划分到指定cost_type下，如果不在，
    则判断其是否在judge_verb中，如果在，则划分到指定judge_verb对应的cost_type下，如果不在
    则继续判断第一个元素是否在cost_type中，如果在，则将该整体list内容划分到指定cost_type下，如果不在，
    则判断其是否在judge_verb中，如果在，则划分到对应的cost_type下，如果不在，则统一归到unknown_cost_type中
    '''
#     if judgement_result is Nan: return {}
    if len(judgement_result) == 0:
        return {}
    if (len(judgement_result) > 1 and type(judgement_result[1])==str):
        result_dict = {}
        if judgement_result[1] in cost_type1:
            result_dict['cost_type1'] = [(judgement_result)]
        elif judgement_result[1] in cost_type2:
            result_dict['cost_type2'] = [(judgement_result)]
        elif judgement_result[1] in cost_type3:
            result_dict['cost_type3'] = [(judgement_result)]
        elif judgement_result[1] in cost_type4:
            result_dict['cost_type4'] = [(judgement_result)]
#         elif judgement_result[1] in cost_type5:
#             result_dict['cost_type5'] = [(judgement_result)]
        elif judgement_result[1] in judge_verb2:
            result_dict['cost_type2'] = [(judgement_result)]
        elif judgement_result[1] in judge_verb3:
            result_dict['cost_type3'] = [(judgement_result)]
        else: 
            result_dict['unknown_cost_type'] = [(judgement_result)]
#         print("judgement_result[1]: ", judgement_result[1], type(judgement_result[1]))
#         print(result_dict)
        if (type(judgement_result[0]==str) and 'unknown_cost_type' in result_dict.keys()):
            result_dict = {}
            if judgement_result[0] in cost_type1:
                result_dict['cost_type1'] = [(judgement_result)]
            elif judgement_result[0] in cost_type2:
                result_dict['cost_type2'] = [(judgement_result)]
            elif judgement_result[0] in cost_type3:
                result_dict['cost_type3'] = [(judgement_result)]
            elif judgement_result[0] in cost_type4:
                result_dict['cost_type4'] = [(judgement_result)]
#             elif judgement_result[0] in cost_type5:
#                 result_dict['cost_type5'] = [(judgement_result)]
            elif judgement_result[0] in judge_verb2:
                result_dict['cost_type2'] = [(judgement_result)]
            elif judgement_result[0] in judge_verb3:
                result_dict['cost_type3'] = [(judgement_result)]
            else: 
                result_dict['unknown_cost_type'] = [(judgement_result)]
#             print("judgement_result[0]: ", judgement_result[0], type(judgement_result[0]))
#             print(result_dict)
        return result_dict
    result_dict = {}
    for i in judgement_result:
        result_dict = merge_dict(result_dict, read_judgement_result(i))
    return result_dict

def split_costtype_one_row(row_in_pd):
    judgere_dict = read_judgement_result(row_in_pd['judgement_result'])
    for cost_type in cost_types:
        row_in_pd[cost_type] = judgere_dict[cost_type] if cost_type in judgere_dict.keys() else {}
    return row_in_pd

def split_cost_type(pd_data):
    return pd_data.apply(lambda row: split_costtype_one_row(row), axis=1)
        
def containsAny(seq, aset):
    return True if any(seq in i for i in aset) else False

def extract_success_rate(row_in_pd):
    row_in_pd['is_success'] = "unknown"
    if row_in_pd['procedure'] != "一审":
        if row_in_pd['judgere_keyword']=="维持": row_in_pd['is_success'] = False
        if row_in_pd['judgere_keyword']=="撤销": row_in_pd['is_success'] = True
    if row_in_pd['procedure'] == "一审":
        if len(row_in_pd['judgement_result']) > 0: 
            if (len(row_in_pd['cost_type2']) == 0 and len(row_in_pd['cost_type3'])==0):
                row_in_pd['is_success'] = False
            else: row_in_pd['is_success'] = True
    return row_in_pd

def extract_success_rate_ins1(pd_data):
    return pd_data.apply(lambda row: extract_success_rate(row), axis=1)
    
```

```python
# 根据百度api为地区添加经纬度信息
# url example: http://api.map.baidu.com/geocoding/v3/?address=北京市海淀区上地十街10号&output=json&ak=您的ak&callback=showLocation
def find_lonlat(row_in_pd):
#     print(list(row_in_pd.index))
    url = 'http://api.map.baidu.com/geocoding/v3/'
    params = { 'address' : row_in_pd['court_name'],           
               'ak' : 'gXNjFLdd8Ot1NzGGy5ZZDrkzuDIvZCt2',          # 百度密钥
               'output': 'json'}          # 输出结果设置为json格式
    res = requests.get(url,params)
    if res.content:
        result = json.loads(res.text)
    #     print("result", result)
        if result['status']==0:
            row_in_pd['lon'] = result['result']['location']['lng']
            row_in_pd['lat'] = result['result']['location']['lat']
    return row_in_pd

#添加经纬度信息
# finaldf = finaldf.apply(lambda row: find_lonlat(row), axis=1)
```

```python
# 更正黑龙江双城区的地理信息
def correct_hlj(row):
        if "双城" in row['court_name']:
            row['province'] = "黑龙江省"
            row['city'] = "哈尔滨市"
            row['district'] = "双城区"
        return row
```

```python
#给带有歧义的法院名称消岐(统一为省+市+区+人民法院的形式)
def merge_courtname(row):
    row['final_court_name'] = row['court_name']
    if row['district'] != '':
        row['final_court_name'] = str(row['province']) + str(row['city']) + str(row['district']) + "人民法院"
    return row

# finaldf_re = finaldf.apply(lambda row: merge_courtname(row), axis=1)
```
