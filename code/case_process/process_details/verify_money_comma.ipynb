```python
def has_comma(row_in_pd):
    has_comma_flag = False
    cost_types = [eval(row_in_pd['cost_type' + str(i)]) for i in range(1, 4) 
                  if row_in_pd['cost_type' + str(i)]!="{}"]
    for cost in cost_types:
        for info in cost:
            if "," in info[2]: 
                has_comma_flag = True
                break
    row_in_pd['has_comma'] = has_comma_flag
    return row_in_pd
total = total.apply(lambda row_in_pd: has_comma(row_in_pd), axis=1)
total = total[total['has_comma']==True]
```

```python
# 重新修正一下all_cases.csv里面的金额
def extract_penalty(row):
    pmoney = [0]
    penalty = eval(row['cost_type3'])
    if len(penalty) > 0:
        for i in penalty:
            money_string = i[2].replace(",", "")
            exresult = c2d.takeNumberFromString(money_string)
            if len(exresult['digitsStringList']) > 0:
                money = exresult['digitsStringList'][0]
                pmoney.append(float(money))
#             row['penalty'] = max(pmoney)
#         else:
#             penalty = eval(row['cost_type5']) #当赔偿金缺失时用共计金额替代，但是还需要优化金额类型的提取
    row['penalty_corrected'] = max(pmoney)
    return row
def extract_objectmoney(row):
    objectmoney = [0]
    penalty = eval(row['cost_type2'])
    if len(penalty) > 0:
        for i in penalty:
            money_string = i[2].replace(",", "")
            exresult = c2d.takeNumberFromString(money_string)
            if len(exresult['digitsStringList']) > 0:
                money = exresult['digitsStringList'][0]
                objectmoney.append(float(money))
    row['objectmoney_corrected'] = max(objectmoney)
    return row
def extract_legalfee(row):
    legalfee = []
    penalty = eval(row['cost_type1'])
    if len(penalty) > 0:
        for i in penalty: 
            money_string = i[2].replace(",", "")
            exresult = c2d.takeNumberFromString(money_string)
            if len(exresult['digitsStringList']) > 0:
                money = exresult['digitsStringList'][0]
                legalfee.append(float(money))
    row['legalfee_corrected'] = min(legalfee) if len(legalfee) > 0 else 0
    return row
```
