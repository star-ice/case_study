```python
# Build Baseline
# Baseline1
all_cases = pd.read_csv('/Users/starice/RA/Legal_case_study/case_study_result/total_extracted_result/all_cases.csv', encoding="utf-8")
# The lawyer inforamtion need to be corrected
all_cases = all_cases[~(all_cases['defendant']==all_cases['lawyer'])]
all_cases['judgement_date'] = pd.to_datetime(all_cases.judgement_date, format='%Y-%m-%d')
```

```python
all_cases.head()
```

```python
previous = all_cases[all_cases['judgement_date'] < datetime(2017, 7, 1)].groupby(['city', 'plaintiff']).agg({'defendant':lambda x: set(x)}).reset_index()
after = all_cases[(all_cases['judgement_date'] >= datetime(2018, 1, 1)) & (all_cases['judgement_date'] < datetime(2020, 1, 1))].groupby(['city', 'plaintiff']).agg({'defendant':lambda x: set(x)}).reset_index()
previous['len'] = previous['defendant'].apply(len)
after['len'] = after['defendant'].apply(len)
```

```python
after.head()
```

```python
baseline1 = after.merge(previous, on=['city', 'plaintiff'], how='left')
baseline1['defendant_y'] = baseline1['defendant_y'].fillna({i: {} for i in baseline1.index})
baseline1['accurate_objects'] = baseline1.apply(lambda r: set(r['defendant_x']).intersection(set(r['defendant_y'])), axis=1)
baseline1['accuracy_num'] = baseline1['accurate_objects'].apply(len)
baseline1['accuracy'] = baseline1['accuracy_num'] / baseline1['len_x']
baseline1['accuracy'].mean(), baseline1['accuracy'].std()
```

```python
# Baseline2
previous = all_cases[all_cases['judgement_date'] < datetime(2017, 7, 1)].groupby(['city', 'plaintiff', 'defendant'])['case_id'].nunique().reset_index().sort_values('case_id', ascending=False)
after = all_cases[(all_cases['judgement_date'] >= datetime(2018, 1, 1)) & (all_cases['judgement_date'] < datetime(2020, 1, 1))].groupby(['city', 'plaintiff', 'defendant'])['case_id'].nunique().reset_index().sort_values('case_id', ascending=False)
# after
```

```python
previous = previous.groupby(['city', 'plaintiff'])['defendant'].apply(list).reset_index()
after = after.groupby(['city', 'plaintiff'])['defendant'].apply(list).reset_index()
after.head()
```

```python
baseline2 = after.merge(previous, on=['city', 'plaintiff'], how="left")
```

```python
def cal_hits(row, num):
        if num > len(row['defendant_x']): num = len(row['defendant_x'])
        row['accurate_objects'] = []
        if len(row['defendant_y']) > 0:
                if len(set(row['defendant_x'][:num]).intersection(set(row['defendant_y']))) > 0:
                        row['accurate_objects'] = list(set(row['defendant_x'][:num]).intersection(set(row['defendant_y'])))
        row['accurate_nums'] = len(row['accurate_objects'])
        return row
def get_hits(baseline2, num):
        baseline2['defendant_y'] = baseline2['defendant_y'].fillna({i: [] for i in baseline1.index})
        baseline2 = baseline2.apply(lambda r: cal_hits(r, num), axis=1)
        baseline2['len_x'] = baseline2['defendant_x'].apply(len)
        baseline2['hits'] = baseline2['accurate_nums'] / baseline2['len_x']
        return baseline2

baseline2_10 = get_hits(baseline2, 10)
baseline2_5 = get_hits(baseline2, 5)
baseline2_3 = get_hits(baseline2, 3)
```

```python
baseline2_10['hits'].mean(), baseline2_10['hits'].std(), baseline2_5['hits'].mean(), baseline2_5['hits'].std(), baseline2_3['hits'].mean(), baseline2_3['hits'].std()
# get_hits(baseline2, 1)['hits'].mean()
# epoch30 hits10: 0.22720191592261904, hits5: 0.17380487351190477, hits3: 0.15476190476190477
# epoch100 hits10: 0.3186668113425926, hits5: 0.251953125, hits3: 0.21269169560185186
```
